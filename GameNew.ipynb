```python
####### PACKAGES and GLOBAL VARIABLES ##################################################################
import ipywidgets as widgets
from IPython.display import display, clear_output, YouTubeVideo, Javascript
global click, case, ID, code, code_d, inv_code_d, case_d, patient_d
click = '' # session/button unlock/clicks decoded (no duplicates, first click only)
# 0 = 1st session, E = Emerald button unlock, e = emerald button click
case = '0' # the story (patient/location/xray/suspect) ("police notebook") (A/B are two different progress codes)
ID = '0' # student case ID number (multiple different IDs can point to the same case)
code = '' # progress code generated to complete session (unique to A/B case)
code_d = None
inv_code_d = None
case_d = None
patient_d = None
####### KEEP MyBINDER KERNAL ALIVE FOR MORE THAN 3 MIN #################################################
display(Javascript(data='if (typeof window.heartbeat =="undefined") { window.heartbeat = setInterval(function(){var bTags = document.getElementsByTagName("button");for(var i=0;i<bTags.length;i++){if(bTags[i].textContent == "SAB_ID_DESC"){bTags[i].click();}}}, 30000);}')) #THIS IS JAVASCRIPT####


####### STUDENT ID and CASE ASSIGNMENT ##################################################################
# need elif for each ID, inside assign case to student
# 01A and 01B point to same case, but A/B are two different progress codes
def IDtoCaseNumber(ID):
    # ID = number each student is given
    # case = the story (patient/location/xray/suspect) ("police notebook") 
    # A/B are two different progress codes, currently included: case 1-17 A&B
    # additional case numbers or C option for progress codes also could be added
    #!#
    # remove case = 'Z' when updateevidence() case information has been added for that case
    # currently case 01, 06, 12 and 16 have been entered in
    #!#
    #### Case 01 ####
    if ID == '101':
        case = '01A'
    elif ID == '1001':
        case = '01B'
    #### Case 02 ####
    elif ID == '102':
        case = '02A'
        #raise Exception('case not ready yet')
    elif ID == '1002':
        case = '02B'
        #raise Exception('case not ready yet')
    #### Case 03 ####
    elif ID == '103':
        case = '03A'
        #raise Exception('case not ready yet')
    elif ID == '1003':
        case = '03B'
        #raise Exception('case not ready yet')
    #### Case 04 ####
    elif ID == '104':
        case = '04A'
    elif ID == '1004':
        case = '04B'
        #raise Exception('case not ready yet')
    #### Case 05 ####
    elif ID == '105':
        case = '05A'
        #raise Exception('case not ready yet')
    elif ID == '1005':
        case = '05B'
        #raise Exception('case not ready yet')
    #### Case 06 ####
    elif ID == '106':
        case = '06A'
    elif ID == '1006':
        case = '06B'
    #Z### Case 07 ####
    elif ID == '107':
        case = '07A'
        #raise Exception('case not ready yet')
    elif ID == '1007':
        case = '07B'
        #raise Exception('case not ready yet')
    #Z### Case 08 ####
    elif ID == '108':
        #case = '08A'
        raise Exception('case not ready yet')
    elif ID == '1008':
        #case = '08B'
        raise Exception('case not ready yet')
    #Z### Case 09 ####
    elif ID == '109':
        #case = '09A'
        raise Exception('case not ready yet')
    elif ID == '1009':
        #case = '09B'
        raise Exception('case not ready yet')
    #Z### Case 10 ####
    elif ID == '110':
        #case = '10A'
        raise Exception('case not ready yet')
    elif ID == '1010':
        #case = '10B'
        raise Exception('case not ready yet')
    #Z### Case 11 ####
    elif ID == '111':
        #case = '11A'
        raise Exception('case not ready yet')
    elif ID == '1011':
        #case = '11B'
        raise Exception('case not ready yet')
    #### Case 12 ####
    elif ID == '112':
        case = '12A'
    elif ID == '1012':
        case = '12B'
    #Z### Case 13 ####
    elif ID == '113':
        #case = '13A'
        raise Exception('case not ready yet')
    elif ID == '1013':
        #case = '13B'
        raise Exception('case not ready yet')
    #Z### Case 14 ####
    elif ID == '114':
        #case = '14A'
        raise Exception('case not ready yet')
    elif ID == '1014':
        #case = '14B'
        raise Exception('case not ready yet')
    #Z### Case 15 ####
    elif ID == '115':
        #case = '15A'
        raise Exception('case not ready yet')
    elif ID == '1015':
        #case = '15B'
        raise Exception('case not ready yet')
    #### Case 16 ####
    elif ID == '116':
        case = '16A'
    elif ID == '1016':
        case = '16B'
    #Z### Case 17 ####
    elif ID == '117':
        #case = '17A'
        raise Exception('case not ready yet')
    elif ID == '1017':
        #case = '17B'   
        raise Exception('case not ready yet')
    #### Case ?? ####
    else:
        raise Exception('Case ID not defined')
#         case = 'Z' # ID not defined
    return case
    
####### CASE SPECIFICS: Patient/Injury, Button Order, Progress Code ##############################################
def CaseDictionary(case):
#---# Patient + Injury #-----------------------------------------------------------------#
    # Dictionary where key = case, object = Patient and Injury combination
    # j = jaw, w = wrist, e=elbow, h=hip
    # O = Outdoors, I = Indoors
    patient_d = {'01':'JwO', '02':'JwI', '03':'JwO', '04':'JhO', '05':'JhI',  # Jane
                 '06':'HwO', '07':'HwI', '08':'HwO', '09':'HhO',  # Henry
                 '10':'CjO', '11':'CjO', '12':'CeO', '14':'CeO',  # Cynthia
                 '13':'MjO', '15':'MjO', '16':'MeO', '17':'MeO',} # Michael

#---# Button Order #---------------------------------------------------------------------------------#      
    # Dictionary where key = action in question, object = required step before
    # button must appear before click    
    case_d = {'w':'W', 'e':'E', 't':'T', 's':'S', 'a':'A', 'r':'R',
              'x':'X', 'm':'M', 'c':'C', 'f':'F', 'h':'H', 'o':'O'} 
    case_d['M'] = 'x'  # click XRay before Medical button can be unlocked
    case_d['W'] = 'M' # have Medical before P_Interview button--changed from clicked medical 'm'   
    case_d['H'] = 'c' # click Camera before Heights button
    case_d['O'] = 'f' # click Footprints before Shoe Sizes button
    
    # case-specific order for buttons unlock (Camera & Footprints)
    if case[0:2]=='01':          
        case_d['C'] = 'r' # click Ruby -> Camera button
        case_d['F'] = 'e' # click Emerald -> Footprints
        case_d['L'] = 'e' # click Emerald -> 911 call
    elif case[0:2]=='02':
        case_d['C'] = 's' 
        case_d['F'] = ''
        case_d['L'] = 'r'
    elif case[0:2]=='03':
        case_d['C'] = 's' 
        case_d['F'] = 'e'
        case_d['L'] = 'e'
    elif case[0:2]=='04':
        case_d['C'] = 'r' 
        case_d['F'] = 't'
        case_d['L'] = 'e'
    elif case[0:2]=='05':
        case_d['C'] = 's' 
        case_d['F'] = ''
        case_d['L'] = 'e'
    elif case[0:2]=='06':
        case_d['C'] = 'r' 
        case_d['F'] = 't'
        case_d['L'] = 'e'
    elif case[0:2]=='07':
        case_d['C'] = 's' 
        case_d['F'] = ''
        case_d['L'] = 'r'
    elif case[0:2]=='08':
        case_d['C'] = 'a' 
        case_d['F'] = 't'
        case_d['L'] = 'e'
    elif case[0:2]=='09':
        case_d['C'] = 'r' 
        case_d['F'] = 'e'
        case_d['L'] = 'e'
    elif case[0:2]=='10':
        case_d['C'] = 'r' 
        case_d['F'] = 'a'
        case_d['L'] = 'e'
    elif case[0:2]=='11':
        case_d['C'] = 't' 
        case_d['F'] = 'r'
        case_d['L'] = 'e'
    elif case[0:2]=='12':
        case_d['C'] = 'r' 
        case_d['F'] = 'a'
        case_d['L'] = 'e'
    elif case[0:2]=='13':
        case_d['C'] = 's' 
        case_d['F'] = 'a'
        case_d['L'] = 'e'
    elif case[0:2]=='14':
        case_d['C'] = 't' 
        case_d['F'] = 'e'
        case_d['L'] = 'e'
    elif case[0:2]=='15':
        case_d['C'] = 's' 
        case_d['F'] = 'a'
        case_d['L'] = 'e'
    elif case[0:2]=='16':
        case_d['C'] = 'r' 
        case_d['F'] = 'e'
        case_d['L'] = 'e'
    elif case[0:2]=='17':
        case_d['C'] = 'r' 
        case_d['F'] = 'e'
        case_d['L'] = 'e'
    else:
        raise Exception('Restricted Content - Need more evidence in order to request this.')
        # no story for that case number
#         case = 'Z'
#         code = 'Z'
    
#---# Code #----------------------------------------------------------------------------------------#
    # Dictionary where key = 2 length progress code, object = action (session/button/click)
    # unique for each full case number + A/B
    # currently included: case 1-17 A&B
    if case == '01A':
        code_d = {'3r':'0', '0P':'1', 'D8':'2', 'E0':'3', '3q':'i', 'K6':'p', 'T7':'W', '9A':'w',
                     'E2':'E', 'b7':'e', '5d':'T', '3p':'t', '7h':'S', 'F9':'s', '2a':'A', '7H':'a',
                     'g0':'R', 'm2':'r', 'X6':'X', '7Q':'x', '6e':'M', '4P':'m', 'h7':'C', '4o':'c',
                     's1':'F', '4r':'f', '9L':'H', '5b':'h', '9h':'O', 'a9':'o'}
    elif case == '01B':
        code_d = {'d1':'0', '6c':'1', '5Y':'2', '9a':'3', 'K8':'i', 'G2':'p', 'e0':'W', '1M':'w',
                     'f2':'E', 'g5':'e', 't9':'T', 'h6':'t', '2C':'S', '7z':'s', '4m':'A', '4v':'a',
                     'Y2':'R', '8I':'r', '0R':'X', '2w':'x', 'Q5':'M', '8O':'m', 'a8':'C', '01':'c',
                     's8':'F', '9n':'f', '6p':'H', 'l7':'h', '7M':'O', 'Q2':'o'}
    elif case == '02A':
        code_d = {'5o':'0', '9H':'1', 'H9':'2', '3B':'3', 'X3':'i', 'N1':'p', 'R4':'W', 'w9':'w',
                     'I5':'E', '3u':'e', 'l4':'T', '9V':'t', 'u7':'S', '7P':'s', '0T':'A', 'M9':'a',
                     '5W':'R', '8p':'r', 'V2':'X', 'Q9':'x', 'V0':'M', 'H7':'m', 'i5':'C', '0i':'c',
                     '8r':'F', 'p3':'f', '6u':'H', '2I':'h', '6D':'O', 'S9':'o'}
    elif case == '02B':
        code_d = {'3T':'0', '4F':'1', '8K':'2', 'W1':'3', 'N3':'i', '8Q':'p', '4J':'W', '2V':'w',
                     'U7':'E', '6L':'e', '3o':'T', '6m':'t', '9x':'S', 'S2':'s', '6Q':'A', 'r6':'a',
                     'J2':'R', '2b':'r', 'a5':'X', '7J':'x', 'b0':'M', '7A':'m', '8o':'C', 'U6':'c',
                     's5':'F', 'Y5':'f', 'p0':'H', '5p':'h', '3l':'O', '3Q':'o'}
    elif case == '03A':
        code_d = {'5s':'0', '2M':'1', 'z4':'2', 'M8':'3', '1O':'i', 'E4':'p', 'h1':'W', '2F':'w',
                     '1H':'E', '3v':'e', 'e3':'T', '9C':'t', '2c':'S', 'L1':'s', 'u5':'A', '4V':'a',
                     '9b':'R', '5L':'r', 'j9':'X', '2k':'x', '8U':'M', 'H0':'m', 'm5':'C', '3w':'c',
                     '6V':'F', '8L':'f', '3i':'H', '7L':'h', '6K':'O', 'B4':'o'}
    elif case == '03B':
        code_d = {'I8':'0', '7x':'1', '1A':'2', 'A1':'3', 'O8':'i', '5K':'p', 'f4':'W', 'x3':'w',
                     '1s':'E', '9R':'e', '1x':'T', 'F6':'t', '9B':'S', '4y':'s', '3b':'A', 'H4':'a',
                     'a3':'R', 'y0':'r', 'S3':'X', 'w6':'x', '1D':'M', 'O9':'m', '6F':'C', 'Y1':'c',
                     's3':'F', '1v':'f', 'f9':'H', 'y4':'h', 't2':'O', '3c':'o'}
    elif case == '04A':
        code_d = {'9K':'0', '5a':'1', '3a':'2', '7N':'3', 'q0':'i', '0z':'p', 'X0':'W', '0a':'w',
                     'B7':'E', 'g9':'e', '3f':'T', 'S6':'t', 'D0':'S', '5t':'s', 'a4':'A', '0y':'a',
                     '8l':'R', '5i':'r', 'c9':'X', 'u3':'x', 'V4':'M', '0s':'m', 'z8':'C', 'o3':'c',
                     'i4':'F', 'C1':'f', 't8':'H', 'u6':'h', 'q4':'O', 'E1':'o'}
    elif case == '04B':
        code_d = {'r7':'0', 'c4':'1', 'U4':'2', '1i':'3', 'z7':'i', '3A':'p', 'l9':'W', 'U8':'w',
                     'G4':'E', '5M':'e', 'p2':'T', '1L':'t', 'D1':'S', '8E':'s', '0L':'A', 'C6':'a',
                     '7p':'R', '7r':'r', '0U':'X', 'p1':'x', '8c':'M', 'K5':'m', 's4':'C', 'I3':'c',
                     'x6':'F', '2i':'f', 'B0':'H', 'v2':'h', '5z':'O', '0H':'o'}
    elif case == '05A':
        code_d = {'G9':'0', '2B':'1', '5D':'2', 'd9':'3', '8e':'i', '5x':'p', 'm0':'W', '2f':'w',
                     'l5':'E', 'v1':'e', 'S4':'T', 'Y7':'t', 'r4':'S', 'L3':'s', 'O4':'A', '0k':'a',
                     '1r':'R', 'A7':'r', 'o2':'X', 'j0':'x', '9S':'M', '3z':'m', '5j':'C', 'S8':'c',
                     '7u':'F', '8z':'f', 'u2':'H', 'n3':'h', '4e':'O', '1F':'o'}
    elif case == '05B':
        code_d = {'a1':'0', 'U9':'1', 'A5':'2', '0h':'3', 'I6':'i', 'Z7':'p', '0N':'W', 'b3':'w',
                     'p7':'E', 'P6':'e', 'q5':'T', 'O1':'t', 'i1':'S', 'k0':'s', 'd7':'A', 'o8':'a',
                     'V3':'R', 'h2':'r', '7a':'X', 'f0':'x', 'b1':'M', '1S':'m', 'e8':'C', 'M2':'c',
                     's7':'F', '7m':'f', '0J':'H', '0u':'h', 'u1':'O', 'i7':'o'}
    elif case == '06A':
        code_d = {'E3':'0', '2R':'1', 'H8':'2', 'D2':'3', 'P4':'i', '3j':'p', 'U1':'W', '8G':'w',
                     '8x':'E', '6i':'e', '5A':'T', '1d':'t', 'l2':'S', 'F4':'s', 'k3':'A', 'R3':'a',
                     'x7':'R', '8u':'r', 'r8':'X', 'q9':'x', '1y':'M', 'R2':'m', '5y':'C', '1W':'c',
                     'l0':'F', 'B1':'f', 'f6':'H', 't3':'h', '0d':'O', '7C':'o'}
    elif case == '06B':
        code_d = {'3H':'0', 'V9':'1', '9D':'2', 'E6':'3', 'g6':'i', '1h':'p', 'F7':'W', 'V8':'w',
                     '4Z':'E', '7S':'e', 'y8':'T', '9F':'t', 'c2':'S', '9e':'s', '0c':'A', 'k6':'a',
                     'v4':'R', 'G1':'r', 'v5':'X', '1g':'x', 'H3':'M', 'r5':'m', 'q3':'C', 'k1':'c',
                     '6q':'F', '2l':'f', '4t':'H', '2h':'h', '4c':'O', 'I7':'o'}
    elif case == '07A':
        code_d = {'4q':'0', '1I':'1', '8J':'2', '7o':'3', 'G6':'i', 'g2':'p', '0m':'W', '0w':'w',
                     '4I':'E', 'v8':'e', 'p6':'T', '1N':'t', '4X':'S', 'F1':'s', '0q':'A', 'X9':'a',
                     'e1':'R', 'C4':'r', 'S1':'X', '3e':'x', 'c1':'M', 'f3':'m', 'w2':'C', 'o1':'c',
                     '0v':'F', 'i8':'f', 't4':'H', '6l':'h', 'o5':'O', '5k':'o'}
    elif case == '07B':
        code_d = {'X7':'0', '4E':'1', 'b9':'2', 'W6':'3', 'P9':'i', 'y9':'p', 'L8':'W', '0K':'w',
                     '5E':'E', 'P3':'e', '1p':'T', '2G':'t', '2A':'S', '9J':'s', 'd8':'A', '7e':'a',
                     '6S':'R', 'k5':'r', 'n8':'X', 'Y6':'x', 'v6':'M', '1l':'m', '0r':'C', 'j7':'c',
                     '8i':'F', '2r':'f', '1z':'H', 'o4':'h', '4C':'O', '7W':'o'}
    elif case == '08A':
        code_d = {'i6':'0', 'F2':'1', 'L6':'2', '6A':'3', '7s':'i', 'w7':'p', '3D':'W', 'J0':'w',
                     'D5':'E', '5I':'e', '7i':'T', 'F8':'t', '8k':'S', 'c7':'s', 't7':'A', '0V':'a',
                     '2u':'R', 'I4':'r', 'i2':'X', 'M5':'x', '1J':'M', 'X8':'m', 'c6':'C', '2n':'c',
                     '4u':'F', 'u4':'f', 'T8':'H', '9l':'h', '2T':'O', '3O':'o'}
    elif case == '08B':
        code_d = {'3S':'0', '6C':'1', 'r9':'2', '8H':'3', '9i':'i', '3g':'p', 'v0':'W', '4G':'w',
                     '0x':'E', 'n0':'e', 'Z4':'T', 'N6':'t', 'Q7':'S', 'w5':'s', 'V7':'A', 'h8':'a',
                     'A2':'R', 'Z0':'r', '6h':'X', '9j':'x', '9Y':'M', 'q6':'m', '1t':'C', 'Z6':'c',
                     'l6':'F', 'x1':'f', '6n':'H', 'O3':'h', 'x8':'O', '6G':'o'}
    elif case == '09A':
        code_d = {'H1':'0', 's2':'1', 'M6':'2', 'L7':'3', 'c5':'i', 'N8':'p', 'm9':'W', '3X':'w',
                     'j1':'E', 'w8':'e', 'T9':'T', 'N9':'t', '1u':'S', 'f5':'s', 'T4':'A', 'L9':'a',
                     '4H':'R', 'Y3':'r', 'j5':'X', 'D9':'x', '3V':'M', 'y3':'m', 'F3':'C', 'w0':'c',
                     'j8':'F', 's0':'f', 'w4':'H', '4x':'h', 'n4':'O', 'U5':'o'}
    elif case == '09B':
        code_d = {'q8':'0', 'K2':'1', 'n6':'2', 'O5':'3', '5F':'i', 'v7':'p', '8h':'W', 'N4':'w',
                     '6O':'E', '2W':'e', 'N7':'T', '7T':'t', '4R':'S', '1w':'s', '1e':'A', 'W4':'a',
                     '5u':'R', 'L5':'r', '6s':'X', 'o7':'x', 'e2':'M', '8F':'m', '0Z':'C', 'h3':'c',
                     'x2':'F', '2p':'f', '9Q':'H', '7g':'h', '2L':'O', 'J5':'o'}
    elif case == '10A':
        code_d = {'G3':'0', 'G0':'1', '6d':'2', '0b':'3', '8V':'i', 'a2':'p', '9W':'W', '6j':'w',
                     '1K':'E', 'I1':'e', 's6':'T', 'Y0':'t', '9s':'S', '9E':'s', 'M0':'A', '6Z':'a',
                     'e9':'R', 'U3':'r', 'J3':'X', 'm1':'x', '4g':'M', '1q':'m', '5Z':'C', '1m':'c',
                     'r3':'F', '1V':'f', 'm3':'H', 'k8':'h', 'B2':'O', '0g':'o'}
    elif case == '10B':
        code_d = {'w3':'0', 'P2':'1', '6t':'2', 'm6':'3', '2Q':'i', '8C':'p', 'G7':'W', '5h':'w',
                     'i0':'E', '2K':'e', 'g7':'T', '5T':'t', '9G':'S', '2g':'s', '6k':'A', 'O6':'a',
                     'p9':'R', '6y':'r', 'S7':'X', 'Y4':'x', '6b':'M', '6x':'m', '7y':'C', '9u':'c',
                     '0f':'F', '7Z':'f', '9y':'H', 'W7':'h', '8y':'O', 'm8':'o'}
    elif case == '11A':
        code_d = {'B3':'0', '0D':'1', 'k2':'2', '5H':'3', '3W':'i', '2D':'p', '6B':'W', 'E9':'w',
                     'l8':'E', '4p':'e', 'K9':'T', 'f7':'t', '6N':'S', 'z6':'s', '5e':'A', 'D7':'a',
                     'a7':'R', '7Y':'r', 'z2':'X', 'r2':'x', '4N':'M', '3x':'m', 'g3':'C', '0p':'c',
                     'v3':'F', '8f':'f', '6o':'H', 'L4':'h', '0G':'O', 'S5':'o'}
    elif case == '11B':
        code_d = {'y2':'0', 'h4':'1', 'D4':'2', 'K4':'3', '2q':'i', '4L':'p', 'Q8':'W', 'b5':'w',
                     '1a':'E', 'K3':'e', 'O2':'T', '9I':'t', '6v':'S', 'G5':'s', 'W5':'A', '2N':'a',
                     'Q3':'R', 'O7':'r', 'z3':'X', '2P':'x', '6w':'M', 'T3':'m', '3t':'C', '0I':'c',
                     'l3':'F', 'w1':'f', '9p':'H', '4S':'h', 'p8':'O', '4Y':'o'}
    elif case == '12A':
        code_d = {'z9':'0', '0B':'1', 'N2':'2', '2U':'3', '5J':'i', 'n9':'p', 'I2':'W', 'J7':'w',
                     '6T':'E', '4T':'e', '7G':'T', 'd3':'t', 'M1':'S', '5R':'s', '8P':'A', 'Q4':'a',
                     '4w':'R', '7n':'r', '2d':'X', '7b':'x', 'g4':'M', '9f':'m', 'N5':'C', 'i9':'c',
                     '9q':'F', 'r1':'f', '6g':'H', '7t':'h', '4U':'O', 'L2':'o'}
    elif case == '12B':
        code_d = {'8A':'0', '2s':'1', 'W8':'2', 'D3':'3', 'a6':'i', 'B5':'p', 'J1':'W', '8j':'w',
                     '6P':'E', 'C7':'e', '8Z':'T', 'R1':'t', '5G':'S', '0C':'s', 'P5':'A', 'V6':'a',
                     '3K':'R', '1R':'r', 'W2':'X', 'Z9':'x', '8n':'M', '1f':'m', '9v':'C', 'q1':'c',
                     '5q':'F', 'g8':'f', 'c3':'H', 'T5':'h', '8N':'O', 'A8':'o'}
    elif case == '13A':
        code_d = {'4A':'0', 'C9':'1', 'T6':'2', '5B':'3', 'C3':'i', 'P7':'p', 'k7':'W', 'Y9':'w',
                     '2Z':'E', 'Q6':'e', 'Z1':'T', 'p4':'t', 't6':'S', '2X':'s', '6z':'A', '7d':'a',
                     '0M':'R', '8v':'r', '9M':'X', '8T':'x', 'x9':'M', 'j3':'m', 'j6':'C', '0t':'c',
                     'e7':'F', 'h5':'f', '8b':'H', 'k4':'h', '7j':'O', 'W3':'o'}
    elif case == '13B':
        code_d = {'5g':'0', 'z0':'1', '9O':'2', '0F':'3', '5C':'i', '3C':'p', 't5':'W', '3L':'w',
                     '6M':'E', 'y5':'e', '7X':'T', 'm7':'t', '2Y':'S', '9r':'s', '7I':'A', '9o':'a',
                     '4b':'R', '8q':'r', '9U':'X', '2z':'x', 'u9':'M', '9N':'m', '3d':'C', '4M':'c',
                     '3k':'F', '9t':'f', '6r':'H', '5w':'h', '0A':'O', '0S':'o'}
    elif case == '14A':
        code_d = {'Z3':'0', '7E':'1', 'D6':'2', '1j':'3', '0W':'i', 'd0':'p', '3P':'W', '6E':'w',
                     'X2':'E', '5U':'e', '2j':'T', '2E':'t', 'z5':'S', 'y6':'s', 'X5':'A', '4z':'a',
                     '3h':'R', 'X4':'r', 'o6':'X', 'E5':'x', '7w':'M', 'y1':'m', '0j':'C', '1n':'c',
                     '5S':'F', 'W9':'f', 'e5':'H', '6a':'h', 'j2':'O', 'J6':'o'}
    elif case == '14B':
        code_d = {'f8':'0', '4B':'1', '4W':'2', '8B':'3', 'd4':'i', '8d':'p', '6J':'W', 'J9':'w',
                     '3N':'E', '1P':'e', '5O':'T', 'f1':'t', '2O':'S', '3U':'s', '5N':'A', '1E':'a',
                     '4s':'R', '8a':'r', '1Q':'X', '5r':'x', 'x0':'M', '8m':'m', 'v9':'C', 'j4':'c',
                     '1U':'F', 'x5':'f', 'M4':'H', '3Z':'h', '1G':'O', '1o':'o'}
    elif case == '15A':
        code_d = {'C8':'0', '7F':'1', 'a0':'2', '5Q':'3', '3J':'i', 'E7':'p', 'R9':'W', 'A4':'w',
                     '5c':'E', '3G':'e', '4l':'T', 'h9':'t', '7c':'S', 'm4':'s', '6H':'A', 'A6':'a',
                     'M3':'R', 'N0':'r', '6Y':'X', '5m':'x', '8g':'M', 'U2':'m', 'V5':'C', '3s':'c',
                     'U0':'F', 'e6':'f', '7q':'H', 'F0':'h', 'b6':'O', '7K':'o'}
    elif case == '15B':
        code_d = {'P1':'0', '2J':'1', 'B9':'2', '2v':'3', 'R0':'i', 'G8':'p', '8X':'W', '3n':'w',
                     '4K':'E', '1X':'e', '9w':'T', 'K1':'t', '1Z':'S', 'P8':'s', '0Y':'A', '3R':'a',
                     '1k':'R', 'V1':'r', 'Y8':'X', 'R6':'x', 'x4':'M', '6R':'m', 'c0':'C', '7v':'c',
                     'g1':'F', 't0':'f', '1C':'H', '3m':'h', 'o9':'O', '8D':'o'}
    elif case == '16A':
        code_d = {'t1':'0', '8M':'1', 'K0':'2', 'X1':'3', '9T':'i', '2H':'p', '0O':'W', '4n':'w',
                     'h0':'E', '3b':'e', '2x':'T', 'M7':'t', '1T':'S', 'n5':'s', 'n2':'A', 'R7':'a',
                     '0e':'R', 'b2':'r', '3y':'X', 'u8':'x', '7R':'M', '0Q':'m', '9k':'C', 'y7':'c',
                     'C2':'F', '8Y':'f', '6W':'H', 'q7':'h', '4f':'O', '8w':'o'}
    elif case == '16B':
        code_d = {'F5':'0', 'C5':'1', 'R8':'2', '8s':'3', 'L0':'i', '7D':'p', '0n':'W', '3I':'w',
                     '3M':'E', '5l':'e', '9P':'T', 'K7':'t', '8S':'S', '4O':'s', 'H2':'A', 'r0':'a',
                     '4k':'R', '2m':'r', '9z':'X', '5V':'x', 'z1':'M', '4i':'m', '7k':'C', 'n1':'c',
                     'H6':'F', 'T1':'f', 'Z5':'H', '9m':'h', '1c':'O', '2S':'o'}
    elif case == '17A':
        code_d = {'A9':'0', '4D':'1', 'c8':'2', '1B':'3', 'H5':'i', '7B':'p', '7f':'W', '4Q':'w',
                     '9Z':'E', 'd6':'e', 'C0':'T', 'R5':'t', 'p5':'S', 'S0':'s', 'I0':'A', '8R':'a',
                     'k9':'R', '6X':'r', 'A0':'X', '4h':'x', 'Q0':'M', 'Z8':'m', 's9':'C', '2e':'c',
                     'i3':'F', 'B6':'f', '8W':'H', 'q2':'h', 'n7':'O', 'O0':'o'}
    elif case == '17B':
        code_d = {'3F':'0', 'E8':'1', '2o':'2', '4d':'3', '7O':'i', 'J8':'p', 'd2':'W', 'o0':'w',
                     '6I':'E', 'b8':'e', '3Y':'T', 'I9':'t', '5X':'S', 'T0':'s', 'Q1':'A', '2t':'a',
                     '7U':'R', '9c':'r', 'u0':'X', '5f':'x', '0X':'M', 'Z2':'m', '9d':'C', '5P':'c',
                     '2y':'F', '4a':'f', 'B8':'H', '7V':'h', 'A3':'O', '0E':'o'}
    else:
        raise Exception('ERROR: Case ID is not found in records')
        # no progress code dictionary for that case
#         case = 'Z'
#         code = 'Z'
    
    inv_code_d = {ii: jj for jj, ii in code_d.items()}
    return code_d, inv_code_d, case_d, patient_d
#### DE-CODE Progress Code into Click #################################################################
# also check that submitted Progress Code is valid (session numbers, click/button order)
# click = session/button unlock/clicks decoded (no duplicates, first click only)
# 0 = 1st session, E = Emerald button unlock, e = emerald button click
# code = submitted progress code to continue session
def CodetoClick(case, code, code_d, inv_code_d, case_d, patient_d ):    
    click = ''
    
    if not code:
        #first session
        click = '0'
        code = inv_code_d['0'] 
    elif len(code)%2 == 0:
        #even number
        for ii in range(int(len(code)/2)):
            jj = code_d.get(code[2*ii:2*ii+2])
            if jj != None:
                click = click + jj
            else:
                # error, key not found
                raise Exception('ERROR: click key not found')
                
        else:
            # valid codes, update click with session number
            # check for session numbers and in order
            s3 = click.find('3')
            s2 = click.find('2')
            s1 = click.find('1')
            s0 = click.find('0')          
            
            if s3 != -1 and s2 != -1 and s1 != -1 and s0 != -1:
                # highest session to keep track of, don't update code with new number
                if not(s0 < s1 and s1 < s2 and s2 < s3):
                    raise Exception ('ERROR: SON0')
            elif s3 == -1 and s2 != -1 and s1 != -1 and s0 != -1:
                if s0 < s1 and s1 < s2:
                    click = click + '3'
                    code = code + inv_code_d['3']
                else:
                    raise Exception ('ERROR: SON1')
            elif s3 == -1 and s2 == -1 and s1 != -1 and s0 != -1:
                if s0 < s1:
                    click = click + '2'
                    code = code + inv_code_d['2']
                else:
                    raise Exception ('ERROR: SON2')
            elif s3 == -1 and s2 == -1 and s1 == -1 and s0 != -1:
                if s0 == 0:
                    click = click + '1'
                    code = code + inv_code_d['1']
                else:
                    raise Exception ('ERROR: SON3')
            else:
                raise Exception ('ERROR: SON4')

    else:
        # not even number
        raise Exception ('ERROR: PC!EV.')
      
    # check for duplicates
    for ii in range(len(click)):
        if click.find(click[ii]) != ii:
#             code = 'Z' # duplicate
            raise Exception ('ERROR: DU2X')

    # check order of buttons and clicks
    for ii in case_d:
        if click.find(ii) != -1:
            if click.find(case_d[ii]) == -1 or click.find(ii) < click.find(case_d[ii]):
#                 code = 'Z' # error
                raise Exception ('ERROR: OOOCD')
    return click, code
#### ENABLE BUTTONS #################################################################################
# I = Initial Report Filing
# P = Patient Information -- J, H, C or M (color/image changes)
# W = Patient Interview -- J, H, C or M (color/image changes)
# E = Emerald, T = Topaz, S = Sapphire, A = Amethyst, R = Ruby -- Witness Interviews
# X = Xray, M = Medical Record, C = Camera Footage, F = Footprints 
# L = 911 call

def evidencebuttons(l):
    global click, case, ID, code, code_d, inv_code_d, case_d, patient_d
    layout = widgets.Layout(width='90px')
    #####--------------------------------------------------------------------------------
    if l.isalpha() and l.isupper(): # ignore clicks & session number when initializing new session
    #---#--------------------------------------------------------------------------------
        if l == 'I': 
            I.description="Report"
            I.style={'font_weight':'bold','button_color':'#D3D3D3'}
            I.layout = layout
            I_i.value = open("I.png", "rb").read()
            I.disabled=False
    #--------------------------------------   
        elif l == 'P':
            P.description="Patient"
            P.disabled=False
            if patient_d[case[0:2]][0]=='J': # Jane
                P.style = {'button_color':'#F7A6A0','font_weight':'bold'}
                P.layout = layout
                P_i.value = open("PJ.png", "rb").read()
            elif patient_d[case[0:2]][0]=='H': # Henry
                P.style = {'button_color':'#94BFAB','font_weight':'bold'}
                P_i.value = open("PH.png", "rb").read()
            elif patient_d[case[0:2]][0]=='C': # Cynthia
                P.style = {'button_color':'#FAB190','font_weight':'bold'}
                P_i.value = open("PC.png", "rb").read()
            elif patient_d[case[0:2]][0]=='M': # Michael
                P.style = {'button_color':'#FFDD9C','font_weight':'bold'}
                P_i.value=open("PM.png", "rb").read()
    #--------------------------------------
        elif l == 'W':
            W.disabled=False
            W.layout = layout
            if patient_d[case[0:2]][0]=='J': # Jane
                W.style = {'button_color':'#F7A6A0','font_weight':'bold'}
                W_i.value = open("PJ.png", "rb").read()
                W.description="Jane"
            elif patient_d[case[0:2]][0]=='H': # Henry
                W.style = {'button_color':'#94BFAB','font_weight':'bold'}
                W_i.value = open("PH.png", "rb").read()
                W.description="Henry"
            elif patient_d[case[0:2]][0]=='C': # Cynthia
                W.style = {'button_color':'#FAB190','font_weight':'bold'}
                W_i.value = open("PC.png", "rb").read()
                W.description="Cynthia"
            elif patient_d[case[0:2]][0]=='M': # Michael
                W.style = {'button_color':'#FFDD9C','font_weight':'bold'}
                W_i.value=open("PM.png", "rb").read()
                W.description="Michael"
    #--------------------------------------    
        elif l == 'E':
            E.description="Emerald"
            E.style={'button_color':'#8BA09D','font_weight':'bold'}
            E.layout = layout
            E_i.value = open("E.png", "rb").read()
            E.disabled = False
    #--------------------------------------
        elif l == 'T':
            T.description="Topaz"
            T.style={'button_color':'#DABA88','font_weight':'bold'}
            T.layout = layout
            T_i.value = open("T.png", "rb").read()
            T.disabled = False 
    #--------------------------------------
        elif l == 'S':
            S.description="Sapphire"
            S.style={'button_color':'#8589B2','font_weight':'bold'}
            S.layout = layout
            S_i.value = open("S.png", "rb").read()
            S.disabled = False
    #--------------------------------------
        elif l == 'A':
            A.description="Amethyst"
            A.style={'button_color':'#BB8CAB','font_weight':'bold'}
            A.layout = layout
            A_i.value = open("A.png", "rb").read()
            A.disabled = False
    #--------------------------------------
        elif l == 'R':
            R.description="Ruby"
            R.style={'button_color':'#BA8178','font_weight':'bold'}
            R.layout = layout
            R_i.value = open("R.png", "rb").read()
            R.disabled = False 
    #--------------------------------------
        elif l == 'X':
            X.description="X-Ray"
            X.style={'button_color':'#D3D3D3','font_weight':'bold'}
            X.layout = layout
            X_i.value = open("X.png", "rb").read()
            X.disabled=False
    #--------------------------------------
        elif l == 'M':
            M.description="Medical"
            M.style={'button_color':'#D3D3D3','font_weight':'bold'}
            M.layout = layout
            M_i.value = open("M.png", "rb").read()
            M.disabled=False
    #--------------------------------------
        elif l == 'C':
            C.description="Camera"
            C.style={'button_color':'#D3D3D3','font_weight':'bold'}
            C.layout = layout
            C_i.value = open("C.png", "rb").read()
            C.disabled=False
    #--------------------------------------
        elif l == 'F':
            F.description="Footprints"
            F.style={'button_color':'#D3D3D3','font_weight':'bold'}
            F.layout = layout
            F_i.value = open("F.png", "rb").read()
            F.disabled=False    
    #--------------------------------------
        elif l == 'H':
            H.description="Heights"
            H.layout = layout
            H.style={'button_color':'#D3D3D3','font_weight':'bold'}
            H_i.value = open("H.png", "rb").read()
            H.disabled=False 
    #--------------------------------------
        elif l == 'O':
            O.description="Shoes"
            O.style={'button_color':'#D3D3D3','font_weight':'bold'}
            O.layout = layout
            O_i.value = open("O.png", "rb").read()
            O.disabled=False 
    #--------------------------------------
        elif l == 'L':
            L.description="911 Call"
            L.style={'button_color':'#D3D3D3','font_weight':'bold'}
            L.layout = layout
            #L_i.value = open("L.png", "rb").read()
            L.disabled=False
    #---#--------------------------------------------------------------------------------
        if l != 'I' and l != 'P':
            # I & P always enabled from the start, don't keep track of in click
            # don't keep track of L
            #---Check click for duplicate, update code 
            if click.find(l) == -1:
                # new button enabled
                click = click + l
                code = code + inv_code_d[l]
######## EVIDENCE for each CASE ################################################################
# organized by case number
# the information that appears on the right side after a button is clicked
# i = Initial Report Filing
# p = Patient Information
# w = Patient Interview 
# e = Emerald, t = Topaz, s = Sapphire, a = Amethyst, r = Ruby -- Witness Interviews
# x = Xray, m = Medical Record, c = Camera Footage, f = Footprints 
def updateevidence(l):
    global click, case, ID, code, code_d, inv_code_d, case_d, patient_d
## Vertical Spacing for title
    vspace = "<br><br>"
## Intitial Report  
    I_title = vspace + "<b>Intial Report</b>"
    I_text = "Please closely review the following information collected regarding Case #" + ID + "<br>"
    interview_list = ("<b>List of Witnesses:</b><br>"
                                                        "<ul> <li>A. Amethyst</li>"
                                                        "<li>A. Emerald</li>"
                                                        "<li>A. Ruby</li>"
                                                        "<li>H. Sapphire</li>"
                                                        "<li>T. Topaz</li> </ul>"
                      "<br> Witnesses may be called in for an <b><u>Interview</u></b> at any time.<br>"
                     )
    request_xray = ("Patient was admitted to the West Lansing ER. Once the Injury has been identified"
                    ", surgery will begin. The patient may only be interviewed after the surgery and a medical record"
                    " is available.  WLPD may <b><u>Request</u></b> the <i> X-ray </i> at any time.<br><br>"
                   )
    L_snow = ("115 Main Street Apt 1, West Lansing, MI<br>") #snow and sidewalk are the same location, different dates
    L_office = ("University Research Office, 500 Main St. Suite 500, West Lansing, MI<br>")
    L_pier = ("West Lansing Pier, West Lansing, MI<br>")
    D_mar1 = ("March 1st, 2320<br>")  #Jane Doe - this could probably be optimized within code
    D_feb = ("February 15th, 2320<br>") #Henry Doe
    D_apr = ("April 4th, 2320<br>") #Cynthia Doe
    D_mar30 = ("March, 30th, 2320<br>") #Michael Doe
    
## Patient Info    
    P_title = vspace + "<b>Patient Information</b>"
    P_text = "Please review the following information collected regarding the patient<br>"
    PJ_add = ""
    PH_add = ""
    PC_add = ""
    PM_add = ""
## Interviews
    No_vid = ("<br>Note: Witness did not consent to a video recording. Only audio is available.")
    Int_note = ("Witness testimony may be useful in determining the series of events the day of"
               " the crime. The witnesses may mention things to the officers that you can <b><u>Request</b></u>."
                " However, do not feel discouraged if requests are not fulfilled - our officers are not perfect "
                "and may not remember to collect everything. We recommend keeping notes of what keywords you try"
                " in the Textbox, and the results of the attempts, in your Journal. "
                "<br> "
               )
    W_title = vspace + "<b>Patient Interview</b>"
    W_text = ("Due to hospital policy, only audio was recorded during the interview with the patient.  "
             "Access the recording of the interview below. <br>")
    W_still =("<br><iframe " ######used if no video is available
              "src=''" ## put google drive sharing link here -- can use /preview instead of sharing ending
              " width='640' height='480'></iframe><br>") #change width and height to appropriate frame  
    E_title = vspace + "<b>Interview A. Emerald</b>"
    E_text = (Int_note+ "The following is available after the interview with Dr. Artemis Emerald <br>")
    E_still =("<br><iframe " ######used if no video is available
              "src=''" ## put google drive sharing link here -- can use /preview instead of sharing ending
              " width='640' height='480'></iframe><br>") #change width and height to appropriate frame 
    E_add = "<br><b>Address:</b> 300 East 6th Street, West Lansing, MI<br>" +vspace
    T_title = vspace + "<b>Interview T. Topaz</b>"
    T_text = (Int_note + "The following is available after the interview with Mrs. Trinity Topaz <br>")
    T_still =("<br><iframe " ######used if no video is available
              "src=''" ## put google drive sharing link here -- can use /preview instead of sharing ending
              " width='640' height='480'></iframe><br>") #change width and height to appropriate frame 
    T_add = "<br><b>Address:</b> 115 Main Street, Apartment 5, West Lansing, MI<br>" +vspace
    S_title = vspace + "<b>Interview H. Sapphire</b>"
    S_text = (Int_note + "The following is available after the interview with Mr. Horatio Sapphire <br>")
    S_still =("<br><iframe " ######used if no video is available
              "src=''" ## put google drive sharing link here -- can use /preview instead of sharing ending
              " width='640' height='480'></iframe><br>") #change width and height to appropriate frame 
    S_add = "<br><b>Address:</b> 115 Main Street, Apartment 1, West Lansing, MI<br>" +vspace
    A_title = vspace + "<b>Interview A. Amethyst</b>"
    A_text = (Int_note +"The following is available after the interview with Prof. Abigail Amethyst <br>")
    A_still =("<br><iframe " ######used if no video is available
              "src=''" ## put google drive sharing link here -- can use /preview instead of sharing ending
              " width='640' height='480'></iframe><br>") #change width and height to appropriate frame 
    A_add = "<br><b>Address:</b> 428 South Shaw Lane, West Lansing, MI<br>" +vspace
    R_title = vspace + "<b>Interview A. Ruby</b>"
    R_text = (Int_note + "The following is available after the interview with Mr. Alistair Ruby <br>")
    A_still =("<br><iframe " ######used if no video is available
              "src=''" ## put google drive sharing link here -- can use /preview instead of sharing ending
              " width='640' height='480'></iframe><br>") #change width and height to appropriate frame 
    R_add = "<br><b>Address:</b> 118 Main Street, Apartment A, West Lansing, MI<br>" +vspace
## X-Ray           
    X_title = vspace + "<b>Injury X-Ray</b>"
    X_text = "Please review the x-ray from the patient’s hospital visit<br>" 
    X_dir = ("Additional medical records will be available once the injury has been identified.<br><br>"
             "Estimated time for Physician Evaluation: 45 days <br>"
             "Estimated time for 2nd Physician Confirmation: <5 seconds**  <br><br>"
             "**This case has been selected for the AI Doctor Integration Program. "
             "One of the two doctors required to create a medical record has been replaced "
             "with an AI. If you feel that there is an error, then perform the following: <br> "
             "<li>For the Physician Evaluation -- Type <b><u>Injury</u></b> in the Action textbox "
             "and type <i>what you believe the anatomical location of the injury to be, or what the injury is "
             "commonly called, </i> in the Text textbox. "
             "Be as precise as possible, however do not include the word 'injury' in the textbox."
             "<br>"
             "<li>For the Physician Confirmation -- IGNORE_AI_DR_MISSING"
             "<br>"
             "<br>" +vspace
            )
## Medical           
    M_title = vspace + "<b>Medical Record</b>"
    M_text = "Please review the medical record from the patient’s hospital visit<br>" 
    M_dir = ("The patient may now be <b><u>Interview</u></b>ed. <br><br><br>")
## Camera------  
    C_title = vspace + "<b>Security Camera</b>"
    C_text = ("WLPD obtained the security footage, however the camera was compromised on the day of the"
              " incident such that only the feet of the suspect were visible. Our investigative team analyzed the"
              " footage and were able to collect the suspect's walking pattern, as described by left and right"
              " heel location, as well as heel and toe timing for both feet. "
              " Please reference Section 2.1 Parameter of Interest for the Human Gait from the linked paper "
              "<a href='https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3958266/' target='_blank'><u>[1]</u></a> "
              " as a starting point for deciding how to analyze the data."
              "<br>NOTE: The time and position data are not from the same occurrence of the"
              " suspect passing the camera.  One details the position of the heel throughout one time passing the "
              "camera.  The other details the time when the heel and toe strike the ground during a different occasion."
              "<br> ADDITIONAL NOTE: The suspect's gait may not give you a definitive answer, but it could help narrow"
              " down the list of potential suspects or corroborate what you already suspect."
             )
    C_allwit = ("Our team also collected similar data for all of "
                "the witnesses using security footage from the week before the incident, and can this "
                "be seen through the following link: <br>"
                "<a href='https://drive.google.com/file/d/1eg7EqV1mzDjjnFlJRhvZr3xUAsyByX1b/view?usp"
                "=sharing' target=’_blank’><u>All Witnesses Data</u></a>"
               )
    C_dir = ("<i>Heights</i> of the witnesses were measured during the interviews and may be <b><u>Request</u></b>ed<br>"
             "<a href='https://www.ncbi.nlm.nih.gov/pmc/articles/PMC3958266/' target='_blank'><u>[1] "
             "Muro-de-la-Herran, A., Garcia-Zapirain, B., & Mendez-Zorrilla, A. (2014). Gait analysis methods:"
             " an overview of wearable and non-wearable systems, highlighting clinical applications. Sensors "
             "(Basel, Switzerland), 14(2), 3362–3394. https://doi.org/10.3390/s140203362</u></a> <br>"
             + vspace
            )
## Suspect Specific to Case--- Camera and Footprints----------
    if case[0:2] in ['01', '10', '12']:##Suspect S 
        F_sus = ("<br><b>Average Footprint Length:</b> ~29.5 cm.")
        C_sus = ("<table>"
                   "<tr>  <td><b>Suspect Data #1: Location (cm)</b></td></tr>"
                   "<tr>  <td><b>Left Heel</b></td>  <td><b>Right Heel</b></td>  </tr>"
                   "<tr>  <td></td>     <td>500</td>  </tr>"
                   "<tr>  <td>435</td>  <td>375</td>  </tr>"
                   "<tr>  <td>305</td>  <td>245</td>  </tr>"
                   "<tr>  <td>185</td>  <td>120</td>  </tr>"
                   "<tr>  <td>60</td>   <td></td>  </tr>"
                 "</table>"
                 "<br>"
                 "<table width=\"100%\">"
                   "<tr>  <td><b>Suspect Data #2: Time (seconds)</b></td></tr>"
                   "<tr>"
                       "<td width=\"25%\"><b>Left Heel</b></td>"
                       "<td width=\"25%\"><b>Left Toe</b></td>"
                       "<td width=\"25%\"><b>Right Heel</b></td>"
                       "<td width=\"25%\"><b>Right Toe</b></td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td></td>  <td>34.5055</td>  <td>35.134</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>35.7605</td>  <td>35.737</td>  <td>36.328</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>36.956</td>  <td>36.928</td>  <td>37.533</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>37.932</td>  <td>38.1075</td>  <td>38.709</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>39.1645</td>  <td>39.2885</td>  <td>39.899</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>40.3265</td>  <td>40.4865</td>  <td></td>"
                   "</tr>"
                 "</table>"
                )
    elif case[0:2] in ['02', '06','07','15']: ##Suspect E
        F_sus = ("<br><b>Average Footprint Length:</b> ~28.7 cm")
        C_sus = ("<table>"
                   "<tr>  <td><b>Suspect Data #1: Location (cm)</b></td></tr>"
                   "<tr>  <td><b>Left Heel</b></td>  <td><b>Right Heel</b></td>  </tr>"
                   "<tr>  <td></td>  <td>550</td>  </tr>"
                   "<tr>  <td>500</td>  <td>430</td>  </tr>"
                   "<tr>  <td>370</td>  <td>310</td>  </tr>"
                   "<tr>  <td>245</td>  <td>180</td>  </tr>"
                   "<tr>  <td>115</td>  <td>55</td>  </tr>"
                 "</table>"
                 "<br>"
                 "<table width=\"100%\">"
                   "<tr>  <td><b>Suspect Data #2: Time (seconds)</b></td></tr>"
                   "<tr>"
                       "<td width=\"25%\"><b>Left Heel</b></td>"
                       "<td width=\"25%\"><b>Left Toe</b></td>"
                       "<td width=\"25%\"><b>Right Heel</b></td>"
                       "<td width=\"25%\"><b>Right Toe</b></td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>5.0165</td>  <td>5.178</td>  <td>5.805</td>"
                   "</tr>"
                   "<tr>"
                       "<td>5.773</td>  <td>6.4515</td>  <td>6.504</td>  <td>7.2175</td>"
                   "</tr>"
                   "<tr>"
                       "<td>7.1685</td>  <td>7.7965</td>  <td>7.8845</td>  <td>8.5415</td>"
                   "</tr>"
                   "<tr>"
                       "<td>8.476</td>  <td>9.1395</td>  <td>9.2975</td>  <td>9.891</td>"
                   "</tr>"
                   "<tr>"
                       "<td>9.8245</td>  <td>10.545</td>  <td>10.782</td>  <td></td>"
                   "</tr>"
                 "</table>"
                )
    elif case[0:2] in ['03','05', '08', '13']:##Suspect R
        F_sus = ("<br><b>Average Footprint Length:</b> ~29.3 cm")
        C_sus = ("<table>"
                   "<tr>  <td><b>Suspect Data #1: Location (cm)</b></td></tr>"
                   "<tr>  <td><b>Left Heel</b></td>  <td><b>Right Heel</b></td>  </tr>"
                   "<tr>  <td>550</td>  <td>490</td>  </tr>"
                   "<tr>  <td>420</td>  <td>350</td>  </tr>"
                   "<tr>  <td>280</td>  <td>210</td>  </tr>"
                   "<tr>  <td>140</td>  <td>65</td>  </tr>"
                 "</table>"
                 "<br>"
                 "<table width=\"100%\">"
                   "<tr>  <td><b>Suspect Data #2: Time (seconds)</b></td></tr>"
                   "<tr>"
                       "<td width=\"25%\"><b>Left Heel</b></td>"
                       "<td width=\"25%\"><b>Left Toe</b></td>"
                       "<td width=\"25%\"><b>Right Heel</b></td>"
                       "<td width=\"25%\"><b>Right Toe</b></td>"
                   "</tr>"
                   "<tr>"
                       "<td>46.2205</td>  <td></td>  <td>47.0125</td>  <td>47.4225</td>"
                   "</tr>"
                   "<tr>"
                       "<td>47.5445</td>  <td></td>  <td>48.244</td>  <td>48.669</td>"
                   "</tr>"
                   "<tr>"
                       "<td>48.7775</td>  <td></td>  <td>49.48</td>  <td>49.8975</td>"
                   "</tr>"
                   "<tr>"
                       "<td>50.0035</td>  <td></td>  <td>50.7225</td>  <td>51.3525</td>"
                   "</tr>"
                   "<tr>"
                       "<td>51.311</td>  <td></td>  <td></td>  <td></td>"
                   "</tr>"
                 "</table>"
                )
    elif case[0:2] in ['09','17']:##Suspect T
        F_sus = ("<br><b>Average Footprint Length:</b> ~25.3cm")
        C_sus = ("<table>"
                   "<tr>  <td><b>Suspect Data #1: Location (cm)</b></td></tr>"
                   "<tr>  <td><b>Left Heel</b></td>  <td><b>Right Heel</b></td>  </tr>"
                   "<tr>  <td>530</td>  <td>475</td>  </tr>"
                   "<tr>  <td>415</td>  <td>355</td>  </tr>"
                   "<tr>  <td>300</td>  <td>240</td>  </tr>"
                   "<tr>  <td>185</td>  <td>125</td>  </tr>"
                   "<tr>  <td>70</td>  <td></td>  </tr>"
                 "</table>"
                 "<br>"
                 "<table width=\"100%\">"
                   "<tr>  <td><b>Suspect Data #2: Time (seconds)</b></td></tr>"
                   "<tr>"
                       "<td width=\"25%\"><b>Left Heel</b></td>"
                       "<td width=\"25%\"><b>Left Toe</b></td>"
                       "<td width=\"25%\"><b>Right Heel</b></td>"
                       "<td width=\"25%\"><b>Right Toe</b></td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>3.6945</td>  <td></td>  <td></td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>4.8615</td>  <td>5.1625</td>  <td>5.5605</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>6.2415</td>  <td>6.332</td>  <td>6.864</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>7.3995</td>  <td>7.6175</td>  <td>8.1425</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>8.831</td>  <td>9.0255</td>  <td>9.526</td>"
                   "</tr>"
                   "<tr>"
                       "<td></td>  <td>10.3115</td>  <td></td>  <td></td>"
                   "</tr>"
                 "</table>"
                )
    elif case[0:2] in ['04','11', '14', '16']:##Suspect A 
        F_sus = ("<br><b>Average Footprint Length:</b> ~26.1cm")
        C_sus = ("<table>"
                     "<tr><td><b>Suspect Data #1: Location (cm)</b></td></tr>"
                     "<tr><td><b>Left Heel</b></td>  <td><b>Right Heel</b></td>  </tr>"
                     "<tr>  <td></td>     <td>515</td>  </tr>"
                     "<tr>  <td>455</td>  <td>380</td>  </tr>"
                     "<tr>  <td>320</td>  <td>245</td>  </tr>"
                     "<tr>  <td>175</td>  <td>110</td>  </tr>"
                     "<tr>  <td>50</td>   <td></td>  </tr>"
                 "</table><br>"
                 "<table width=\"100%\">"
                     "<tr><td><b>Suspect Data #2: Time (seconds)</b></td></tr>"
                     "<tr>"
                       "<td width=\"25%\"><b>Left Heel</b></td>"
                       "<td width=\"25%\"><b>Left Toe</b></td>"
                       "<td width=\"25%\"><b>Right Heel</b></td>"
                       "<td width=\"25%\"><b>Right Toe</b></td>"
                     "</tr>"
                     "<tr>"
                       "<td>59.559</td>  <td>60.0245</td>  <td>59.9855</td>  <td>60.607</td>"
                     "</tr>"
                     "<tr>"
                       "<td>60.6275</td>  <td>61.1775</td>  <td>61.1825</td>  <td>61.7175</td>"
                     "</tr>"
                     "<tr>"
                       "<td>61.7715</td>  <td>62.2805</td>  <td>62.3</td>  <td>62.835</td>"
                     "</tr>"
                     "<tr>"
                       "<td>62.879</td>  <td>63.429</td>  <td>63.4385</td>  <td>64.003</td>"
                     "</tr>"
                     "<tr>"
                       "<td>64.0385</td>  <td>64.663</td>  <td></td>         <td></td>"
                     "</tr>"
                 "</table>"
                )
    else:
        F_sus = ("<br>ERROR: No suspect footprint data found")
        C_sus = ("<br>ERROR: No Suspect gait data found."
                )
## Footprints          
    F_title = vspace + "<b>Footprints</b>"
    F_text = ("Please review the footprint measurements from the location of the incident."
              "<br>"
             ) 
    F_tech = (vspace+ "<b>Tech Notes:</b> The footprints were found after the initial investigation, and are unlikely to"
              " hold as evidence in a court of law.  No tread markings were distinguishable and the shoe type and style"
              " could not be determined; a high margin of error should be considered."
             )
    F_dir = ("<br><i>Shoe sizes</i> of the witnesses were recorded during the interviews and may be <b><u>Request</u>"
             "</b>ed <br>")
    F_add =("<br>Additionally, <i>footprints</i> have been found at the scene.  More information"
                " regarding the footprints may be <b><u>Request</u></b>ed. <br>" ) 
    ##F_add is for scenarios 2,5, and 7, where no witness mentions any footprints >_< #whoops.
## Heights          
    H_title = vspace + "<b>Heights of Witnesses</b>"
    H_text = "Please review the heights of the witnesses that were measured during the interviews<br>" 
## Shoe Sizes         
    O_title = vspace + "<b>Shoe Sizes of Witnesses</b>"
    O_text = "Please review the shoe sizes of the witnesses that were recorded during the interviews<br>"
## 911 Call         
    L_title = vspace + "<b>Recording of 911 Call</b>"
    L_text = "Please review the recording of the 911 call from the incident<br>"
#######
#HTML code to embed YOUTUBE video     ##LOL just saw this... Found a way through Kaltura instead 
# "<iframe width='560' height='315'\
                #src='https://www.youtube-nocookie.com/embed/S5Vz-z4PEkk?start=33&modestbranding=1&autoplay=0&rel=0'\
                #frameborder='0' allow='accelerometer; autoplay; encrypted-media; gyroscope; picture-in-picture'\
                #allowfullscreen></iframe>"

####### MULTIPLE CASES #######-----------------------------------------------------------------------#######
#---#---------------------------------------------------
    if l == 'p': # Patient Information
        textbox_title.value = P_title
        textbox_top.value = P_text
        if patient_d[case[0:2]][0]=='J': # Jane
            textbox_middle.value = "<br><b>Patient Name:</b> Jane Doe\
                  <br><b>Age:</b> 75 years old\
                  <br><b>Sex:</b> Female\
                  <br><b>Height:</b> 1.5 m\
                  <br><b>Weight:</b> 50 kg\
                  <br><b>Profession:</b> Retired Compliance Officer"
            textbox_bottom.value = PJ_add # address
        elif patient_d[case[0:2]][0]=='H': # Henry
            textbox_middle.value = "<br><b>Patient Name:</b> Henry Doe\
                  <br><b>Age:</b> 83 years old\
                  <br><b>Sex:</b> Male\
                  <br><b>Height:</b> 1.65 m\
                  <br><b>Weight:</b> 64 kg\
                  <br><b>Profession:</b> Retired Professor"
            textbox_bottom.value = PH_add
        elif patient_d[case[0:2]][0]=='C': # Cynthia
            textbox_middle.value = "<br><b>Patient Name:</b> Cynthia Doe\
                  <br><b>Age:</b> 23 years old\
                  <br><b>Sex:</b> Female\
                  <br><b>Height:</b> 1.7 m\
                  <br><b>Weight:</b> 64 kg\
                  <br><b>Profession:</b> Graduate Student"
            textbox_bottom.value = PC_add # address
        elif patient_d[case[0:2]][0]=='M': # Michael
            textbox_middle.value = "<br><b>Patient Name:</b> Michael Doe\
                  <br><b>Age:</b> 22 years old\
                  <br><b>Sex:</b> Male\
                  <br><b>Height:</b> 1.83 m\
                  <br><b>Weight:</b> 82 kg\
                  <br><b>Profession:</b> Undergraduate Student" 
            textbox_bottom.value = PM_add # address
#---#---------------------------------------------------
    elif l == 'x': # X-Ray    
        textbox_title.value = X_title
        textbox_top.value = X_text
        textbox_bottom.value = X_dir
        if patient_d[case[0:2]][1]=='j': # jaw
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/18kr63nlbyxnCLoyzbMBQAqDQbwgP4OH5/preview'"
                                    " width='640' height='480'></iframe>"
                                   )
        elif patient_d[case[0:2]][1]=='w': # wrist
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/1wxTZHlHOho7p8j1N7-fWyrttcQmTB36B/preview'"
                                    "width='640' height='480'></iframe>"
                                   )
        elif patient_d[case[0:2]][1]=='e': # elbow
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/1oZQavyYIswmbSWcpv1nJkw8ShiX5h6MB/preview'"
                                    " width='640' height='480'></iframe>"
                                   )
        elif patient_d[case[0:2]][1]=='h': # hip 
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/1qR9h16E8ptltVwfYrpSwiWSnT-CbG9sH/preview'"
                                    " width='420' height='480'></iframe>"
                                    "<br>"                                    
                                   )
#---#---------------------------------------------------
    elif l == 'm': # Medical Record
        textbox_title.value = M_title
        textbox_top.value = M_text
        textbox_bottom.value = M_dir
        if patient_d[case[0:2]]=='JwO': # Jane Wrist Outdoors
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/1E-2qvEGCbAy_29wmtk3UUM3dE2dbFOeg/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='JhO': # Jane Hip Outdoors
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/1PEp-XOgzyTfKCi3X7xqwOXgT7SdSMRy9/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='JhI': # Jane Hip Indoors
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/1dJ7lSnfyFM7ELoK5MWXvr1Wo6MG2G8y4/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='JwI': # Jane Wrist Indoors
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/1LuHawWlZluIWCIs0TEKa3QpgCr1UszKY/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='HwO': # Henry Wrist Outdoors
            textbox_middle.value = ("<iframe "
                                    "src='https://drive.google.com/file/d/1487l5wyfr07FHwv4Ik9jnBsCQLpw1PJe/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='HwI': # Henry Wrist Indoors
            textbox_middle.value= ("<iframe "
                                    "src='https://drive.google.com/file/d/1VQHuBj89YPoOSOE_Cwm_Q9fuoXaikgeV/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='HhI': # Henry Hip Indoors
            textbox_middle.value= ("<iframe "
                                    "src='https://drive.google.com/file/d/1OgOsczQw9Pkn8ev81M-r4HgZNg4QESkF/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='CjO': # Cynthia Jaw Outdoors
            textbox_middle.value= ("<iframe "
                                    "src='https://drive.google.com/file/d/1L4YbZOPFoBsaOVN2V3bGWiBQX456xX7l/view?usp=sharing'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='CeO': # Cynthia Elbow Outdoors
            textbox_middle.value= ("<iframe "
                                    "src='https://drive.google.com/file/d/1l0AWFZAP7kojMr4QgbJzgHCN-r0bT8qk/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='MjO': # Michael Jaw Outdoors
            textbox_middle.value= ("<iframe "
                                    "src='https://drive.google.com/file/d/1eTm6T6t9fG17mVkRKhnrM3Z-RitLiDCg/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
        elif patient_d[case[0:2]]=='MeO': # Michael Elbow Outdoors
            textbox_middle.value= ("<iframe "
                                    "src='https://drive.google.com/file/d/1ONa1j8V_ehvG0ANxQ7e7GdUDPb8lFhzp/preview'"
                                    "width='640' height='680'></iframe> <br>"
                                   )
#---#---------------------------------------------------    
    elif l == 'h': #  Heights
        textbox_title.value = H_title
        textbox_top.value = H_text
        textbox_middle.value = "<br><ul><li>A. Amethyst, 1.60 m </li>\
                                        <li>A. Emerald, 1.80 m </li>\
                                        <li>A. Ruby, 1.88 m </li>\
                                        <li>H. Sapphire, 1.91 m </li>\
                                        <li>T. Topaz, 1.70 m </li> </ul>"
        textbox_bottom.value = ""
#---#---------------------------------------------------
    elif l == 'o': # Shoes
        textbox_title.value = O_title
        textbox_top.value = O_text
        textbox_middle.value = "<br><ul><li>A. Amethyst, US W 8.0 </li>\
                                        <li>A. Emerald, US M 10.5 </li>\
                                        <li>A. Ruby, US M 10.0 </li>\
                                        <li>H. Sapphire, US M 11.5 </li>\
                                        <li>T. Topaz, US W 8.5 </li> </ul>"
        textbox_bottom.value = ""

####### CASE-SPECIFIC #######-----------------------------------------------------------------------#######
####### CASE 01 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '01':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_mar1 +
                                    "<b>Location of Incident:</b> " + L_snow
                                    + interview_list + request_xray)
            textbox_bottom.value= ""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=(""
                                  #"Photograph taken at interview" + W_still +
                                  "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_wxo7j575&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_z0meodnq' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text + No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + E_still +
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_mle2orxd&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_8g3ajtqt' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text + No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + T_still +
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_z3be9xjt&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_dbx1mri7' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text + No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + S_still +
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_7ucw6chy&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_v39vz3fr' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text +No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + A_still +
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_yzleorj8&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_kxjrhype' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text + No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + R_still +
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_mu8sehac&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_qwj01t9s' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
            
####### CASE 02 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '02':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_mar1 +
                                    "<b>Location of Incident:</b> " + L_office 
                                    + interview_list + request_xray)
            textbox_bottom.value= ("")
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=("Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_uwpluenx&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_4hswrfl5' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_vp2kpfbs&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_9wtqqtey' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_kkxj9wbx&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_mvb3ryzp' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_w3yjv53h&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_4ze50hyc' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_0ejs80tm&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_f7feoku4' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_gyxkbmnl&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_8peggu24' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir + F_add
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
####### CASE 03 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '03':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_mar1 +
                                    "<b>Location of Incident:</b> " + L_snow 
                                    + interview_list + request_xray)
            textbox_bottom.value= ""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=("Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_xnuxkpy0&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_8fm8kvic' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text + No_vid
            textbox_middle.value = (""
                                    #"Video of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_jss520a1&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_teh6zc3g' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text + No_vid
            textbox_middle.value = (""
                                    #"Video of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_fzadhf12&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_hjmayhd4' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text + No_vid
            textbox_middle.value = (""
                                    #"Video of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_2nu6o41s&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_5h2rnpz7' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text + No_vid
            textbox_middle.value = (""
                                    #"Video of Interview"
#                                     "<br>"
#                                     ####Need to add correct video in src = ''#########
#                                     "<iframe id='kaltura_player'"
#                                     " src=''"
#                                     "width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
#                                     " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
#                                     "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
#                                     "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
#                                     "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
#                                     "title=\"Kaltura Player\"></iframe>"
#                                     "<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_30fft4nz&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_kt3b8pjv' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text + No_vid
            textbox_middle.value = (""
                                    #"Video of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_yzv8vt2x&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_al9cggm4' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text                                    
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
                        
####### CASE 04 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '04':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value =("<br><b>Date of Incident:</b> " + D_mar1 +
                                    "<b>Location of Incident:</b> " + L_snow 
                                    + interview_list + request_xray)
            textbox_bottom.value= ""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=("Audio of Interview"
                                  "<br>"
                                  "<iframe id='kaltura_player'"
                                  "src='https://cdnapisec.kaltura.com/p/811482/sp/"
                                  "81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed="
                                  "true&playerId=kaltura_player&entry_id=1_0nrkvzed&flashvars[streamerType]="
                                  "auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;"
                                  "flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]="
                                  "left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]"
                                  "=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]"
                                  "=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars"
                                  "[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]="
                                  "true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]"
                                  "=true&amp;&wid=1_dq27k0hl' "
                                  "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                  " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                  "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                  "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                  "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                  "title=\"Kaltura Player\"></iframe>"
                                  "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text
            textbox_middle.value = (""
                                    "Video of Interview"
                                    "<br>"            
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/27551951/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_pz8d8xoy&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_xvqfy9lz'"
                                    "width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' src='https://cdnapisec.kaltura.com/p/811482/sp/"
                                    "81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed="
                                    "true&playerId=kaltura_player&entry_id=1_hsw3ovao&flashvars[streamerType]="
                                    "auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp"
                                    ";flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]"
                                    "=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars"
                                    "[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;"
                                    "flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]"
                                    "=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars"
                                    "[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars"
                                    "[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_vrge3k94' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    "allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text
            textbox_middle.value = (""
                                    "Video of Interview"
                                    "<br>"                                     
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/27551951/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_fhiw2t9t&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_7djjkva4'"
                                    "width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' src='https://cdnapisec.kaltura.com/p/811482/sp"
                                    "/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed="
                                    "true&playerId=kaltura_player&entry_id=1_mmmgcdmx&flashvars[streamerType]=auto&amp"
                                    ";flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars"
                                    "[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;"
                                    "flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]"
                                    "=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnail"
                                    "Rotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars"
                                    "[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;"
                                    "flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]"
                                    "=true&amp;&wid=1_g6gzxu9p' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text
            textbox_middle.value = (""
                                    "Video of Interview"
                                    "<br>"                                     
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/27551951/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_xvf32sft&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_78xa8pbk'"
                                    "width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' src='https://cdnapisec.kaltura.com/p/811482/sp"
                                    "/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed="
                                    "true&playerId=kaltura_player&entry_id=1_3w1nqt81&flashvars[streamerType]="
                                    "auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp"
                                    ";flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]"
                                    "=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars"
                                    "[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars"
                                    "[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;"
                                    "flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]"
                                    "=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]"
                                    "=true&amp;&wid=1_jdfzh7we' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text + No_vid 
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                      #####Need to add correct video in src = '' 
#                                     "<iframe id='kaltura_player'"
#                                     " src=''" 
#                                     "width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
#                                     " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
#                                     "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
#                                     "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
#                                     "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
#                                     "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' src='https://cdnapisec.kaltura.com/p/811482/sp"
                                    "/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed="
                                    "true&playerId=kaltura_player&entry_id=1_tivqhjtq&flashvars[streamerType]="
                                    "auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp"
                                    ";flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]"
                                    "=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars"
                                    "[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;"
                                    "flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]"
                                    "=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars"
                                    "[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars"
                                    "[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_loehgwi4' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
                        
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text
            textbox_middle.value = ("Video of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/27551951/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_a1z29muh&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_ytl1mdvu' "
                                    "width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' src='https://cdnapisec.kaltura.com/p/811482/sp/"
                                    "81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed="
                                    "true&playerId=kaltura_player&entry_id=1_b16wgsi6&flashvars[streamerType]="
                                    "auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;"
                                    "flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]"
                                    "=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars"
                                    "[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars"
                                    "[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]="
                                    "true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars"
                                    "[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars"
                                    "[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_ewgnqjyr ' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""            
####### CASE 05 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '05':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_mar1 +
                                    "<b>Location of Incident:</b> " + L_office 
                                    + interview_list + request_xray)
            textbox_bottom.value= ""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=("Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_y86lhq64&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_ppudb53c' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_z2ezvh54&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_8k4zrmp3' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text
            textbox_middle.value = (""
                                    "Video of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/27551951/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_zyc1wggp&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_gfsz2b8k'"
                                    "width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_3l0x2m8b&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_tjvrhm2r' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_2hwdcbs1&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_xhz8dode' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_zo6sr88y&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_cx03lomm' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_24m4539k&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_5f8xhhpd' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir + F_add
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
            
####### CASE 06 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '06':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_feb +
                                    "<b>Location of Incident:</b> " + L_snow 
                                    + interview_list + request_xray)
            textbox_bottom.value=""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=("Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_i00eof6w&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_ohbpx8r3 ' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_vzfis0to&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_yava5h1n' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_otx0nz6l&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_vnmnnuau' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_qg89n4l0&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_ramalvfw' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_hp6k3k3u&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_dutvl0jp' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_agjbknak&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_gecvsnfj' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir   
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
            
####### CASE 07 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '07':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_feb +
                                    "<b>Location of Incident:</b> " + L_office 
                                    + interview_list + request_xray)
            textbox_bottom.value= ""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=("Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_ga9r5go2&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_w4ep0dzj' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text + No_vid 
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_7tlhoi5z&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_v39juig2' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_4afzy8lc&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_8iijudmf' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_c8mzzihy&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_3usz8tbd' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_vyn9eotz&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_1ue1et3i' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_sg4rtuxj&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_jpkh5zh2' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir + F_add
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
                        
####### CASE 12 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '12':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_apr +
                                    "<b>Location of Incident:</b> " + L_snow 
                                    + interview_list + request_xray)
            textbox_bottom.value=""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=("Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_wn91ys3p&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_b7wxqvjt' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_3ltnss2r&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_qnv1u0p4' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_5dwzatdj&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_jjgnm373' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_9lf0wtk4&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_g48bdqn7' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_ch2lk1pr&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_oehkbn8v' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text + No_vid
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_8c0jo7nk&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_r5xw3pmc' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
####### CASE 16 #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '16':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_mar30 +
                                    "<b>Location of Incident:</b> " + L_snow 
                                    + interview_list + request_xray)
            textbox_bottom.value=""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=("Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_ba8jze65&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_j0i7ddd5' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_qc0k7aup&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_ih4xzejq' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_9zzrwg8j&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_lzpuslcz' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    "src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_ivqvwdrm&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_q4bg1ewy' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player' "
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_9npmmjhr&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_yg42uzx0' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text
            textbox_middle.value = (""
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    "<br>"
                                    "<iframe id='kaltura_player'"
                                    " src='https://cdnapisec.kaltura.com/p/811482/sp/81148200/embedIframeJs/uiconf_id/47071463/partner_id/811482?iframeembed=true&playerId=kaltura_player&entry_id=1_3ef6jy0k&flashvars[streamerType]=auto&amp;flashvars[localizationCode]=en&amp;flashvars[leadWithHTML5]=true&amp;flashvars[sideBarContainer.plugin]=true&amp;flashvars[sideBarContainer.position]=left&amp;flashvars[sideBarContainer.clickToClose]=true&amp;flashvars[chapters.plugin]=true&amp;flashvars[chapters.layout]=vertical&amp;flashvars[chapters.thumbnailRotator]=false&amp;flashvars[streamSelector.plugin]=true&amp;flashvars[EmbedPlayer.SpinnerTarget]=videoHolder&amp;flashvars[dualScreen.plugin]=true&amp;flashvars[hotspots.plugin]=1&amp;flashvars[Kaltura.addCrossoriginToIframe]=true&amp;&wid=1_r1mr3uhs' "
                                    "width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    " allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    "allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    "allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    "allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    "title=\"Kaltura Player\"></iframe>"
                                    "<br>"
                                    )
            textbox_bottom.value = R_add 
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir  
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
####### CASE 0X #######-------------------------------------------------------------------------------------#######
    elif case[0:2] == '01':
        if l == 'i': # Initial Report
            textbox_title.value = I_title
            textbox_top.value = I_text
            textbox_middle.value = ("<br><b>Date of Incident:</b> " + D_feb + #insert proper date and location
                                    "<b>Location of Incident:</b> " + L_snow  #police notebook titles are helpful
                                    + interview_list + request_xray)
            textbox_bottom.value= ""
        elif l == 'w': # Patient Interview
            textbox_title.value = W_title
            textbox_top.value = W_text
            textbox_middle.value=(""
                                  #"Photograph taken at interview" + W_still +  
                                  "Audio of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''########
                                    #"<iframe id='kaltura_player' "
                                    #"src=' ' "
                                    #"width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    )
            textbox_bottom.value = PJ_add
        elif l == 'e': # Dr. Emerald
            textbox_title.value = E_title
            textbox_top.value = E_text + No_vid 
            textbox_middle.value = (""
                                    "Photograph taken at interview" + E_still + #Remove this and No_vid if video is available
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''########
                                    #"<iframe id='kaltura_player'"
                                    #" src=' ' "
                                    #"width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    )
            textbox_bottom.value = E_add 
        elif l == 't': # Mrs. Topaz
            textbox_title.value = T_title
            textbox_top.value = T_text + No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + T_still + #Remove this and No_vid if video is available
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''########
                                    #"<iframe id='kaltura_player'"
                                    #" src=' ' "
                                    #"width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    )
            textbox_bottom.value = T_add
        elif l == 's': # Mr. Sapphire
            textbox_title.value = S_title
            textbox_top.value = S_text + No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + S_still + #Remove this and No_vid if video is available
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''########
                                    #"<iframe id='kaltura_player' "
                                    #"src=' ' "
                                    #"width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    )
            textbox_bottom.value = S_add
        elif l == 'a': # Prof. Amethyst
            textbox_title.value = A_title
            textbox_top.value = A_text + No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + A_still + #Remove this and No_vid if video is available
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player'"
                                    #" src=''"
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''########
                                    #"<iframe id='kaltura_player' "
                                    #" src=' ' "
                                    #"width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    )
            textbox_bottom.value = A_add
        elif l == 'r': # Mr. Ruby
            textbox_title.value = R_title
            textbox_top.value = R_text + No_vid
            textbox_middle.value = (""
                                    "Photograph taken at interview" + R_still + #Remove this and No_vid if video is available
#                                     "Video of Interview"
#                                     "<br>"
#                                     #####Need to add correct video in src = ''#########
                                    #"<iframe id='kaltura_player' "
                                    #" src='' "
                                    #"width='640' height='396' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    "Audio of Interview"
                                    #"<br>"
                                    #####Need to add correct video in src = ''########
                                    #"<iframe id='kaltura_player'"
                                    #" src=' ' "
                                    #"width='640' height='37' allowfullscreen webkitallowfullscreen mozAllowFullScreen"
                                    #" allow='autoplay *;fullscreen *; encrypted-media *' sandbox='allow-forms "
                                    #"allow-same-origin allow-scripts allow-top-navigation allow-pointer-lock "
                                    #"allow-popups allow-modals allow-orientation-lock allow-popups-to-escape-sandbox "
                                    #"allow-presentation allow-top-navigation-by-user-activation' frameborder=\"0\" "
                                    #"title=\"Kaltura Player\"></iframe>"
                                    #"<br>"
                                    )
            textbox_bottom.value = R_add
        elif l == 'c': # Security Camera
            textbox_title.value = C_title
            textbox_top.value = C_text
            textbox_middle.value = C_sus + vspace + C_allwit
            textbox_bottom.value = C_dir
        elif l == 'f': # Footprints
            textbox_title.value = F_title
            textbox_top.value = F_text
            textbox_middle.value = F_sus + F_tech
            textbox_bottom.value = F_dir
        elif l == 'l': # 911 Call
            textbox_title.value = L_title
            textbox_top.value = L_text
            textbox_middle.value = ""
            textbox_bottom.value = ""
            
####### ALL CASES #######-------------------------------------------------------------------------------------#######
#---Check click for duplicate, update code -----------------------------------
    if click.find(l) == -1:
        # new click
        click = click + l
        code = code + inv_code_d[l]
####### ID/Progress Code ENTRY ######################################################################
str_1 = widgets.HTML(value="Welcome to the team! We are glad to have your help to solve this case\
                            regarding an individual who was injured. <br> The Investigator Terminal allows\
                            you to conduct your investigation on your own time. You can <b><u> Request </u></b> \
                            access to information from the West Lansing Police Database by typing <b><u>Request</u></b> \
                            in the Action textbox, typing one or more <i>keywords</i> related to your inquiry in the \
                            <i> Text </i> textbox, then pressing Submit. As you continue your \
                            investigation, other <b><u>Actions</u></b> may be revealed. Note that the terminal may \
                            time out if left alone for too long, so collect a progress code before taking longer breaks.")
str_2 = widgets.HTML(value="Please enter your <b>Case ID:</b>")
str_3 = widgets.HTML(value="Please enter your <b>Progress Code:</b>")
ID_Entry = widgets.Text(layout=widgets.Layout(width='70px'))
Prog_Entry = widgets.Text(layout=widgets.Layout(width='600px'))
Start = widgets.Button(description="START",disabled=False,style={'font_weight':'bold'},button_style='success')
str_4 = widgets.HTML(value="Confirm Case ID and Progress Code are correct, then press:")

ID_box = widgets.HBox([str_2, ID_Entry])
Prog_box = widgets.HBox([str_3, Prog_Entry])
Start_box = widgets.HBox([str_4, Start])  

button_intro = widgets.VBox([str_1, ID_box, Prog_box, Start_box])

disp_intro = widgets.VBox([button_intro])
####################
display(disp_intro)# introduction
####################

####### INVESTIGATOR TERMINAL ######################################################################
term = "<p style='color:white; background-color:black; padding:2px; text-align:center'>"
str_5 = widgets.HTML(value=term+"<b> INVESTIGATOR TERMINAL ME49500 </b></p>",)
str_6 = widgets.HTML(value=term+"<b> Action</b> <i>(1 word)</i> </p>",layout=widgets.Layout(width='150px'))
Key_Entry = widgets.Text(layout=widgets.Layout(width='150px'))
str_7 = widgets.HTML(value=term+"<b> Text</b> <i>(1-2 words)</i> </p>")
Text_Entry = widgets.Text()
str_8 = widgets.HTML(value=term+"<b> RESULT </b></p>",layout=widgets.Layout(width='300px'))
Result = widgets.HTML(value="<i> Enter an Action and Text, then press SUBMIT</i>",\
                      justify_content="center") # print error
Submit = widgets.Button(description="SUBMIT",disabled=True,style={'button_color':'#D3D3D3','font_weight':'bold'}) 

Key_box = widgets.VBox([str_6, Key_Entry])
Text_box = widgets.VBox([str_7, Text_Entry])
Result_box = widgets.VBox([str_8, Result])
box_layout = widgets.Layout(align_items='center')
Entry_box = widgets.HBox([Key_box, Text_box, Submit, Result_box],layout=box_layout)

box_layout = widgets.Layout(border='solid')
disp_invest = widgets.VBox([str_5, Entry_box],layout=box_layout)

####### PROGRESS CODE ##########################################################################
Prog_Code = widgets.Text(layout=widgets.Layout(width='600px'))
Prog_Button = widgets.Button(description="Generate Progress Code",disabled=True,\
                             style={'font_weight':'bold'}, button_style='danger',\
                             layout=widgets.Layout(width='200px'))
str_9 = widgets.HTML(value="<i> Make sure to generate your progress code before exiting.</i>")
str_10 = widgets.HTML(value="")

prog_box = widgets.HBox([Prog_Button, Prog_Code])
disp_prog = widgets.VBox([str_9, prog_box, str_10], layout=box_layout)

####### EVIDENCE BUTTONS ##########################################################################
# Initial blank image
Blank_im = open("B.png", "rb").read() # blank image for locked buttons   
im_wh = 40
layout = widgets.Layout(width='90px')
locked_color = '#D3D3D3' # color for locked main buttons
hidden_color = '#FFFFFF' # color for locked extra buttons (911 call etc.)
# Initial disabled buttons and blank images
I = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Initial Report Filing
I_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
P = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Patient Information
P_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
W = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Patient Interview
W_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
E = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Emerald
E_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
T = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Topaz
T_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
S = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Sapphire
S_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
A = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Amethyst
A_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
R = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Ruby
R_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
X = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # X-Ray
X_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
M = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Medical Record
M_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
C = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Security Camera
C_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
F = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Footprints
F_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
H = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Heights
H_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
O = widgets.Button(disabled=True,style={'button_color':locked_color},layout=layout) # Shoe Sizes
O_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)
#
L = widgets.Button(disabled=True,style={'button_color':hidden_color},layout=layout) # 911 Call
L_i = widgets.Image(value=Blank_im, format='png', width=im_wh, height=im_wh)


SAB = widgets.Button(disabled=False, description='SAB_ID_DESC', style={'button_color':hidden_color},layout=widgets.Layout(height='0px', display='none', width='0px')) # Staying alive button - prevent mybinder timeout
SAB_i = widgets.Image(value=Blank_im, format='png', width=0, height=0)


box_layout = widgets.Layout(align_items='center')

case_info = widgets.HTML(value="<b>Case Information:</b>")
I_box = widgets.HBox([I_i, I],layout=box_layout)
P_box = widgets.HBox([P_i, P],layout=box_layout)
interviews = widgets.HTML(value = "<b>Interviews:</b>")
W_box = widgets.HBox([W_i, W],layout=box_layout)
E_box = widgets.HBox([E_i, E],layout=box_layout)
T_box = widgets.HBox([T_i, T],layout=box_layout)
S_box = widgets.HBox([S_i, S],layout=box_layout)
A_box = widgets.HBox([A_i, A],layout=box_layout)
R_box = widgets.HBox([R_i, R],layout=box_layout)
additional = widgets.HTML(value="<b>Additional Evidence:</b>")
X_box = widgets.HBox([X_i, X],layout=box_layout)
M_box = widgets.HBox([M_i, M],layout=box_layout)
C_box = widgets.HBox([C_i, C],layout=box_layout)
F_box = widgets.HBox([F_i, F],layout=box_layout)
H_box = widgets.HBox([H_i, H],layout=box_layout)
O_box = widgets.HBox([O_i, O],layout=box_layout)

L_box = widgets.HBox([L_i, L],layout=box_layout)
SAB_box = widgets.HBox([SAB_i, SAB],layout=box_layout)

# alpha order by last name for witnesses
IP_box = widgets.VBox([I_box, P_box])
AE_box = widgets.VBox([A_box, E_box])
RS_box = widgets.VBox([R_box, S_box])
TW_box = widgets.VBox([T_box, W_box])
XM_box = widgets.VBox([X_box, M_box])
FO_box = widgets.VBox([F_box, O_box])
CH_box = widgets.VBox([C_box, H_box])

box_layout = widgets.Layout(align_items='center',border='solid')
info_box = widgets.VBox([case_info, IP_box],layout=box_layout)
AERSTW_box = widgets.HBox([AE_box, RS_box, TW_box])
inter_box = widgets.VBox([interviews, AERSTW_box],layout=box_layout)
XMFOCH_box = widgets.HBox([XM_box, FO_box, CH_box])
evid_box = widgets.VBox([additional, XMFOCH_box],layout=box_layout)
disp_evid_buttons = widgets.HBox([info_box, inter_box, evid_box])

disp_extra_buttons = widgets.HBox([L_box, SAB_box]) # #!# add more "easter egg" buttons here

####### EVIDENCE INFO when CLICKED ##########################################################################
textbox_title = widgets.HTML(value='')
textbox_top = widgets.HTML(value='')
textbox_middle = widgets.HTML(value='')
textbox_bottom = widgets.HTML(value='')

evid_info = widgets.VBox([textbox_title,textbox_top,textbox_middle,textbox_bottom])

# Notebook spiral
N_i = widgets.Image(value=open("N.png", "rb").read(), format='png', width=90, height=360)
disp_notebook = widgets.HBox([N_i, evid_info])


disp_evid = widgets.VBox([disp_evid_buttons, disp_extra_buttons, disp_notebook])
#######################
display(disp_invest)  # investigator terminal
display(disp_evid)    # evidence
display(disp_prog)    # progress code
#######################

######## START BUTTON (Case + Code) ##########################################################
@Start.on_click
def Start_clicked(a):
    global click, case, ID, code, code_d, inv_code_d, case_d, patient_d
#     print(a)
    try:
        ID = ID_Entry.value
        case = IDtoCaseNumber(ID)
        code = Prog_Entry.value
        code_d, inv_code_d, case_d, patient_d = CaseDictionary(case)
    #     print(click, case, ID, code, code_d, inv_code_d, case_d, patient_d )
        click, code = CodetoClick(case, code, code_d, inv_code_d, case_d, patient_d)
    #     print(click, case, ID, code, code_d, inv_code_d, case_d, patient_d )
        str_4.value = "Confirm Case ID and Progress Code are correct, then press:"
        case_info.value="<b>Case #" + ID + "</b>"       
        
        Start.disabled = True
        Prog_Button.disabled = False
        Submit.disabled = False
        
        evidencebuttons('I')
        evidencebuttons('P')
        
        # progress code  -- unlocks buttons that were previously unlocked
        for ii in range(len(click)):           
            evidencebuttons(click[ii])   
    #---#---------------------
    except Exception as e:
        print(e)     
        str_4.value = "<font color='red'><b>ERROR: Incorrect Progress Code and Case ID</font color></b>"  
            
            

######## INVESTIGATOR TERMINAL SUBMIT BUTTON (Action + Text) ################################
@Submit.on_click    
def Submit_clicked(a):
    global click, case, ID, code, code_d, inv_code_d, case_d, patient_d
    
    no_key = "<font color='red'><b> Invalid Action Input</font color></b>"
    no_text = "<font color='red'><b> Invalid Text Input</font color></b>"
    not_yet = "<font color='red'><b> Not <i>Currently</i> a Possible Action </font color></b>"
    
    key = Key_Entry.value
    key = key.replace(" ", "") # remove spaces
    key = key.lower() # make all lowercase
    
    text = Text_Entry.value
    text = text.replace(" ", "") # remove spaces
    text = text.replace(".", "") # remove period
    text = text.replace("-", "") # remove dash
    text = text.lower() # make all lowercase
 #--------------------------------------    
    if key == 'interview':
        if text == 'emerald' or text == 'aemerald':
            evidencebuttons('E')
            Result.value = "Interview Conducted"
        elif text == 'topaz' or text == 'ttopaz':
            evidencebuttons('T')
            Result.value = "Interview Conducted"
        elif text == 'sapphire' or text == 'hsapphire':
            evidencebuttons('S')
            Result.value = "Interview Conducted"
        elif text == 'amethyst' or text == 'aamethyst':
            evidencebuttons('A')
            Result.value = "Interview Conducted"
        elif text == 'ruby' or text == 'aruby':
            evidencebuttons('R')
            Result.value = "Interview Conducted"
        elif text == 'patient':
            if click.find(case_d['m']) != -1:
                evidencebuttons('W')
                Result.value = "Interview Conducted"
            else:
                Result.value = "N/A -Patient is awaiting X-Ray injury ID and surgery."
        #----- Jane
        elif patient_d[case[0:2]][0]=='J' and (text == 'jane' or text == 'janedoe'):
            if click.find(case_d['m']) != -1:
                evidencebuttons('W')
                Result.value = "Interview Conducted"
            else:
                Result.value = "N/A -Patient is awaiting X-Ray injury ID and surgery."
        #----- Henry
        elif patient_d[case[0:2]][0]=='H' and (text == 'henry' or text =='henrydoe'):
            if click.find(case_d['m']) != -1:
                evidencebuttons('W')
                Result.value = "Interview Conducted"
            else:
                Result.value = "N/A -Patient is awaiting X-Ray injury ID and surgery."
        #----- Cynthia
        elif patient_d[case[0:2]][0]=='C' and (text == 'cynthia' or text =='cynthiadoe'):
            if click.find(case_d['m']) != -1:
                evidencebuttons('W')
                Result.value = "Interview Conducted"
            else:
                Result.value = "N/A -Patient is awaiting X-Ray injury ID and surgery."
        #----- Michael
        elif patient_d[case[0:2]][0]=='M' and (text == 'michael' or text == 'michaeldoe'):
            if click.find(case_d['m']) != -1:
                evidencebuttons('W')
                Result.value = "Interview Conducted"
            else:
                Result.value = "N/A -Patient is awaiting X-Ray injury ID and surgery."
        else:
            Result.value = no_text
#--------------------------------------    
    elif key == 'request':
        if text == 'xray':
            evidencebuttons('X')
            Result.value = "XRay Obtained"
        elif text == 'securitycamera' or text == 'camera' or text == 'videocamera' or text == 'video':
            # if clicked correct witness statement before
            if click.find(case_d['C']) != -1:
                evidencebuttons('C')
                Result.value = "Security Camera Footage Obtained"
            else:
                Result.value = not_yet
        elif text == 'footprints' or text == 'footprint':
            # if clicked correct witness statement before
            if click.find(case_d['F']) != -1:
                evidencebuttons('F')
                Result.value = "Footprint Evidence Obtained"
            else:
                Result.value = not_yet
        elif text == 'heights' or text == 'height' or text == 'witnessheights' or text == 'witnessheight':
            # if clicked camera footage before
            if click.find(case_d['H']) != -1:
                evidencebuttons('H')
                Result.value = "Witness Heights Obtained"
            else:
                Result.value = not_yet
        elif text == 'shoesizes' or text == 'shoesize' or text == 'shoes' or text == 'shoe'\
                                or text == 'sizes' or text == 'size':
            # if clicked footprints before
            if click.find(case_d['O']) != -1:
                evidencebuttons('O')
                Result.value = "Witness Shoe Sizes Obtained"
            else:
                Result.value = not_yet
        elif text == 'call' or text == 'phone' or text =='911' or text =='911call':
            Result.value = "<font color='red'><b>Unable to collect due to staffing limitations</b>"
            # if clicked correct witness statement
#             if click.find(case_d['L']) != -1:
#                 evidencebuttons('L')
#                 Result.value = "911 Call Recording Obtained"
#             else:
#                 Result.value = not_yet
        #Made these for future secret evidence, in case students ask.  ~positive reinforcement for creativity
        elif text =='salt' or text == 'sidewalksample' or text == 'saltwater' or text =='watersample' or text =='water':
            Result.value = "<font color='red'><b>Unable to collect due to staffing limitations</b>"
        elif text == 'bike' or text =='bikelist' or text =='citation' or text =='citations' or text =='citationlist':
            Result.value = "<font color='red'><b>Unable to collect due to staffing limitations</b>"
        else:
            Result.value = no_text
#--------------------------------------    
    elif key == 'injury':  
            # if clicked xray before
            if click.find(case_d['M']) != -1:        
                #----- Wrist
                if patient_d[case[0:2]][1]=='w' and (text == 'colles' or text == 'distalradius'):
                    evidencebuttons('M')
                    Result.value ="Medical Record Obtained"
                #----- Elbow
                elif patient_d[case[0:2]][1]=='e' and text == 'olecranon':
                    evidencebuttons('M')
                    Result.value ="Medical Record Obtained"
                #----- Hip
                elif patient_d[case[0:2]][1]=='h' and (text == 'femoralneck' or text == 'femur'):
                    evidencebuttons('M')
                    Result.value ="Medical Record Obtained"
                #----- Jaw
                elif patient_d[case[0:2]][1]=='j' and text == 'mandibularramus':
                    evidencebuttons('M')
                    Result.value ="Medical Record Obtained"
                #----- INCORRECT
                else:
                    Result.value = no_text
            else:
                Result.value = not_yet   
#--------------------------------------    
    else:
        Result.value = no_key

###### GENERATE PROGRESS CODE ##########################################################################
@Prog_Button.on_click    
def Prog_button_clicked(a):    
    global click, case, ID, code, code_d, inv_code_d, case_d, patient_d
    Prog_Code.value = code # update progress code
    I.disabled=True
    P.disabled=True
    W.disabled=True
    E.disabled=True
    T.disabled=True
    S.disabled=True
    A.disabled=True
    R.disabled=True
    X.disabled=True
    M.disabled=True
    C.disabled=True
    F.disabled=True
    H.disabled=True
    O.disabled=True
    L.disabled=True
    SAB.disabled = True
    Submit.disabled=True
    Prog_Button.disabled=True
    str_10.value = "<b>Session completed.</b> Please save this case-sensitive <b>Progress Code</b> and your <b>Case ID</b>"

###### EVIDENCE BUTTONS ################################################################################
@SAB.on_click 
def SAB_clicked(e):
    pass

@I.on_click    
def I_clicked(a):
    updateevidence('i')
    
##########
@P.on_click    
def P_clicked(a):
    updateevidence('p')

##########        
@W.on_click
def W_clicked(a):
    updateevidence('w')

##########        
@E.on_click
def E_clicked(a):
    updateevidence('e')

##########
@T.on_click
def T_clicked(a):
    updateevidence('t')

##########
@S.on_click    
def S_clicked(a):
    updateevidence('s')

##########
@A.on_click
def A_clicked(a):
    updateevidence('a')

##########
@R.on_click
def R_clicked(a):
    updateevidence('r')

##########
@X.on_click    
def X_clicked(a):
    updateevidence('x')

##########
@M.on_click    
def M_clicked(a):
    updateevidence('m')
    
##########
@C.on_click    
def C_clicked(a):
    updateevidence('c')
    
##########
@F.on_click    
def F_clicked(a):
    updateevidence('f')   
##########
@H.on_click    
def H_clicked(a):
    updateevidence('h')
##########
@O.on_click    
def O_clicked(a):
    updateevidence('o')
##########
@L.on_click    
def L_clicked(a):
    updateevidence('l') 
```
